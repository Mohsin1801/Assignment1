```python
def isPrime(num):
    for n in range(2,int(num**1/2)+1):
        if num%n==0:
            return False
    return True

digit = 0
sum = 0
rem = 0
counter = 0
i = 0
checkNum = 0
length = 0
j = 0
r = 10 ** 2 

# Here 'r' will be equal to 100.
# This program is generic for how many range you are giving. Here, for 1 Million iterations it is not running it will be
# so much time complexity.

for j in range(2,r):

    checkNum = j
    sum = 0
    i = 0
    length = len(str(checkNum))
    
    # This loop is for total number of roatations exists for that number.
    while (i < length):
        rem = checkNum % 10
        checkNum = int(checkNum / 10)
        checkNum = (rem * (10 ** (length - 1)) + checkNum)

        digit = isPrime(checkNum)
        sum = sum + digit
        i += 1

        if (sum == length):
            counter += 1

print('There are {0} Circular Prime Numbers below {1}.'.format(counter,r))
```
