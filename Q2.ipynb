```python
def isPrime(num):
    
    if num == 0 or num == 1:
        return False
    else:
        
        for n in range(2,int(num**1/2)+1):
            if num%n==0:
                return False
        return True

i = 0
primes = []

while True:
    if isPrime(i):
        if sum(primes) + i > 1000000:
            break
        
        primes.append(i)
    
    i += 1
    
max = 0
total = 0
    
for i in range(len(primes)):
    for j in range(i,len(primes) + 1):
        if isPrime(sum(primes[i:j])):
            if j - i > max:
                max = j - i
                total = sum(primes[i:j])
                
print(max)
print(total)
```
