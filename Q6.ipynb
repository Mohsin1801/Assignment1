```python
class CircularQueue:
  
    def __init__(self, maxSize):
        self.queue = list()
    
        self.maxSize = maxSize
        self.head = -1
        self.tail = -1
        self.count = 0

    def isFull(self):
        if self.size() == self.maxSize:
            return True
        else:
            return False
    
    def isEmpty(self):
        if self.size() == 0:
            return True
        else:
            return False
        
    def enqueue(self, data):
    
        if self.isFull():
            print("\n< Queue is full! >")
        else:
            self.queue.append(data)
            self.tail = (self.tail+1) % self.maxSize
            self.count += 1
        return True
  
    def dequeue(self):
    
        if self.isEmpty():
            print("\n< Queue is empty! >")
        else:
            data = self.queue[self.head]
            self.head = (self.head+1) % self.maxSize
            self.count -= 1
            return data
    
    def dequeueLeft(self):
        if self.isEmpty():
            print('\n< Queue is empty! >')
        else:
            data = self.queue[self.tail]
            self.tail = (self.head + 1) & self.maxSize
            self.count -= 1
            return data
    
    def size(self):
        return self.count

class Stack:
 
    def __init__(self):
        self.q1 = CircularQueue(5)
        self.q2 = CircularQueue(5)
    
    def push(self, data):
    
        if self.q1.isEmpty():
            self.q1.enqueue(data)
        else:
            q1Size = self.q1.size()
            
            for i in range(0,q1Size):
                self.q2.enqueue(self.q1.dequeue)
            
            self.q1.enqueue(data)
            
            for j in range(0,q1Size):
                self.q1.enqueue(self.q2.dequeue)
    
    def pop(self):
        elem = self.q1.dequeue()
        
        return elem
        

keys = [1, 2, 3, 4, 5]
 
s = Stack()

for key in keys:
    s.push(key)
    
while s:
    print(s.pop())
    
print(s.pop())
```
