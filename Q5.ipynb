```python
class CircularQueue:
  
    def __init__(self, maxSize):
        self.queue = list()
    
        self.maxSize = maxSize
        self.head = -1
        self.tail = -1
        self.count = 0

    def isFull(self):
        if self.size() == self.maxSize:
            return True
        else:
            return False
    
    def isEmpty(self):
        if self.size() == 0:
            return True
        else:
            return False
        
    def enqueue(self, data):
    
        if self.isFull():
            print("\n< Queue is full! >")
        else:
            self.queue.append(data)
            self.tail = (self.tail+1) % self.maxSize
            self.count += 1
        return True
  
    def dequeue(self):
    
        if self.isEmpty():
            print("\n< Queue is empty! >")
        else:
            data = self.queue[self.head]
            self.head = (self.head+1) % self.maxSize
            self.count -= 1
            return data
    
    def size(self):
        return self.count
  
    def FindMin(self):
        minimum = 0
    
        if(self.isEmpty()):
            print('\n< Queue is Empty! >')
        else:
            list = self.queue
            minimum = min(list)
            return minimum
        return True

size = input("Enter the size : ")
q = CircularQueue(int(size))

value = 0
userOption = 0

while True:
    print('1 to Enqueue')
    print('2 to Dequeue')
    print('3 to get Size')
    print('4 to check isFull')
    print('5 to check isEmpty')
    print('6 to Display')
    print('7 to Find Minimum')
    print('0 to Stop')
    userOption = int(input('Enter Choice : '))
    
    if userOption == 1:
        value = input('Enter Values to insert : ')
        q.enqueue(value)
    
    elif userOption == 2:
        q.dequeue()
    
    elif userOption == 3:
        print(q.size())
        
    elif userOption == 4:
        if q.isFull():
            print('\n< Queue is Full! >')
        else:
            print('\n< Queue is Not Full! >')
    elif userOption == 5:
        if q.isEmpty():
            print('\n< Queue is Empty >')
        else:
            print('\n< Queue is not Empty >')
    elif userOption == 6:
        print(q.queue)
    elif userOption == 7:
        print('\n',q.FindMin(),' is the Minimum Number in Queue ')
    elif userOption == 0:
        break
    else:
        print('Invalid Input\n')
        break
    print('\n')

q.FindMin()
```
