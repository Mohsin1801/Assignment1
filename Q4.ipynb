```python
def isPallindrome(n):
    temp = n
    reverse =0
    
    while(n>0):
        digit = n % 10
        reverse = reverse * 10 + digit
        n = n // 10
    if(temp == reverse):
        return True
    else:
        return False

def rev_add(x):
    return int(str(x)[::-1]) + x

iter_LIMIT = 50
LIMIT = 10000
counter = 0

for i in range(LIMIT):
    num = i
    for j in range(iter_LIMIT):
        num = rev_add(num)
        if isPallindrome(num) == 1:
            break
        if j==49:
            counter+=1

print('There are ',counter,'Lychrel numbers below ',LIMIT,'.')
```
