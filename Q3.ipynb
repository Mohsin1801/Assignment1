```python
def isPrime(num):
    
    if num == 0 or num == 1:
        return False
    else:
        
        for i in range (2,int(num ** 1/2) + 1):
            if num % i == 0:
                return False
        return True

def isTruncatable(num):
    
    length = len(str(num))
    magicNum1 = 10              # it will be used for the truncation of both left and right
    magicNum2 = 10 ** (length - 1)
    conv_num = 0
    flag1 = False
    
    for i in range(length - 1):
        conv_num = num // magicNum1
        
        if isPrime(conv_num):
            magicNum1 *= 10
            flag1 = True
        else:
            flag1 = False
            break
    
    if flag1 == False:
        return flag1
    else:
        for i in range(length - 1):
            conv_num = num % magicNum2
        
            if isPrime(conv_num):
                magicNum2 /= 10
                flag2 = True
            else:
                flag2 = False
                break

    return (flag1 == True and flag2 == True)
    
primes = []
count = 0
i = 0

while (count != 11):
    
    if isPrime(i):
        if isTruncatable(i):
            primes.append(i)
            count += 1
    
    i += 1

# it will show 11 truncatable primes 
print(primes)
print(sum(primes))
```
